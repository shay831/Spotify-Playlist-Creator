```python
# need to make sure each full name is just one element
# do this by first removing 'SONGS', 'STREAMS', and number of streams

def ismil(string):
    """Returns true if the string represents millions"""
    string = string[:-1] if string[-1] == 'M' else ""
    string = string.replace('.','') if string.count('.') == 1 else string
    string = string.isdigit()
    return string
    
top500 = list(filter(lambda s: not ismil(s) and s != 'SONG' and s != 'STREAMS' and s != 'NEW!', top500))
```

```python
def nameLstToStr(lst):
    s = ""
    for n in lst:
        s = s + n + ' '
    return s[:-1]

topnames = []
rank = 0
for i, el in enumerate(top500):
    if el.isdigit() and int(el) == rank + 1:
        rank = int(el)
        name = nameLstToStr(top500[i+1:top500.index(str(rank+1)) if rank < 500 else None])
        topnames.append(name)

        
```

## Getting All Song IDs of Top 500 Artists

```python
def songids(ide):
    """Returns a list of all song ids of a particular artist."""
    songids = []
    albs = sp.artist_albums(ide)['items']
    for i in range(len(albs)):
        albid = albs[i]['id']
        songs = sp.album_tracks(albid)['items']
        for j in range(len(songs)):
            songids.append(songs[j]['id'])
        
    return list(set(songids))
```

```python
songs = []
for i, el in enumerate(topids):
    songs = songs + songids(el)
    print("-", end = '') if i % 10 == 9 else None
```

## Pulling Data from Artists

There are a few statistics on artists that are not listed in the song data that could potentially be good features. These are genres, followers, and Spotify's popularity metric. Since some songs have featured artists, we will get the followers and popularity of both the main artist and the total of all artists on the song. We will only look at genres of the main artist.

```python
# create dictionary with artist ID as key and tuple of genres, followers, and popularity as value
art_data = {}
for k in topids:
    a = sp.artist(k)
    art_data[str(k)] = (a['genres'], a['followers']['total'], a['popularity'])
```

```python
art_keys = set(art_data.keys())
artists_data_mat = []
for i, s in enumerate(song_info):
    if s['popularity'] >= 30:
        artists = s['artists']
        main = artists[0]
        if main['id'] in art_keys:
            main_genres = art_data[main['id']][0]
            main_fols = art_data[main['id']][1]
            main_pop = art_data[main['id']][2]
        else:
            main_genres = sp.artist(main['id'])['genres']
            main_fols = sp.artist(main['id'])['followers']['total']
            main_pop = sp.artist(main['id'])['popularity']

        tot_fols = 0
        tot_pop = 0
        for a in artists:
            if a['id'] in art_keys:
                tot_fols += art_data[a['id']][1]
                tot_pop += art_data[a['id']][2]
            else:
                tot_fols += sp.artist(a['id'])['followers']['total']
                tot_pop += sp.artist(a['id'])['popularity']

        row = [s['id'], main_fols, main_pop, tot_fols, tot_pop, main_genres]
        artists_data_mat.append(row)
    print("-", end = '') if (i+1) % 1000 == 0 else None
```

```python
artist_cols = ['id', 'main_fols', 'main_pop', 'tot_fols', 'tot_pop', 'main_genres']
artist_data = pd.DataFrame(artists_data_mat, columns=artist_cols)
```

```python
# list of all genres
genres = []
for k in topids:
    for g in art_data[k][0]:
        if g not in genres:
            genres.append(g)
```

```python
def manyhot(genre_lists, unique_genres):
    df = pd.DataFrame()
    for g in unique_genres:
        cat = [int(g in i) for i in genre_lists]
        df[g] = cat
        df[g] = cat
    return df

genre_data = manyhot(artist_data['main_genres'], genres)
```
